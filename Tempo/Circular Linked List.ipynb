```python
class CLL_Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CLL_Construction:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = CLL_Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def remove(self):
        if self.head is None:
            return "Empty List"
        if self.head.next == self.head:
            removed_data = self.head.data
            self.head = None
            return removed_data
        temp = self.head
        while temp.next.next != self.head:
            temp = temp.next
        removed_data = temp.next.data
        temp.next = self.head
        return removed_data

    def search(self, key):
        if self.head is None:
            return None
        temp = self.head
        while True:
            if temp.data == key:
                return temp
            temp = temp.next
            if temp == self.head:
                break
        return None
    
    def traverse(self):
        if self.head is None:
            print("The list is empty.")
            return
        temp = self.head
        while True:
            print(temp.data, end=" --> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(Back to Head)")
```

```python
def main():
    clist = CLL_Construction()
    print("*** Circular Linked List ***")
    while True:
        print("\n1. Insert Element")
        print("2. Remove Element")
        print("3. Search Element")
        print("4. Display List")
        print("5. Exit")
        
        choice = int(input("Enter your choice: "))
        if choice == 1:
            value = eval(input("Enter the value to be inserted: "))
            clist.insert(value)
            print(f"{value} inserted into the list.")
        elif choice == 2:
            removed_value = clist.remove()
            if removed_value == "Empty List":
                print("Cannot delete. The list is empty.")
            else:
                print(f"Value Deleted: {removed_value}")
        elif choice == 3:
            value = eval(input("Enter the value to search: "))
            result = clist.search(value)
            if result:
                print(f"Element {value} found in the list.")
            else:
                print(f"Element {value} not found.")
        elif choice == 4:
            clist.traverse()
        elif choice == 5:
            print("Exiting program. Goodbye! (❁´◡`❁)")
            break
        else:
            print("Invalid Choice. Please try again.")

if __name__ == "__main__":
    main()
```
